```python
import pandas as pd
import numpy as np
```

```python
u_cols = ['movieId', 'title', 'genres']
data = pd.read_csv('movies.csv',header=0,names=u_cols)
```

```python
data.head()
```

```python
# Esto es un filtro
movies = data[data.movieId <= 10]
```

```python
movies
```

```python
movies.genres.str.get_dummies(sep='|')
```

```python
dataset = pd.concat([data, data.genres.str.get_dummies(sep='|')], axis=1)
```

```python
dataset.head()
```

```python
movie_categories = dataset.columns[3:].tolist()
```

```python
movie_categories
```

```python
dataset.tail()
```

```python
dataset[dataset.movieId == 25].title
```

```python
dataset.loc[24]
```

```python
def get_user_preferences(user_movies):
    user_features = dataset[dataset.title.isin(user_movies)].iloc[:,3:].T
    user_features = user_features.mean(axis=1).reset_index()
    return user_features.iloc[:,1].tolist()
```

```python
user_preferences_list = get_user_preferences(user_movies)
user_preferences_list
```

```python
def get_predicted_movie_score(movie_name, user_preferences):
    movie_slice = dataset[dataset.title==movie_name].iloc[0]
    movie_features = movie_slice[movie_categories]
    return np.dot(movie_features, user_preferences)
```

```python
get_predicted_movie_score('Armageddon (1998)', user_preferences_list)
```

```python
get_predicted_movie_score('Lambert & Stamp (2014)', user_preferences_list)
```

```python
#Adventure #recomendable
get_predicted_movie_score('The Great Alone (2015)', user_preferences_list)
```

```python
def get_movie_recommendations(user_preferences, number_recommendations):
    dataset['score'] = dataset.title.apply(get_predicted_movie_score, args=([user_preferences]))
    dataset.sort_values(by=['score'], ascending=False, inplace=True)
    return dataset[['title','genres','score']].head(number_recommendations)
```

```python
get_movie_recommendations(user_preferences_list, 10)
```
